```python
# plot iterations vs cumulative reward
def plot_reward(df_stats):
    plt.plot(df_stats['Reward'])
    plt.title('Iterations vs Reward')
    plt.xlabel('number of Iterations')
    plt.ylabel('Reward')
    return plt

#plot_reward(df_stats)
```

```python
# plot iterations vs delta error
def plot_error(df_stats):
    plt.plot(df_stats['Error'])
    plt.title('Iterations vs Delta Error')
    plt.xlabel('number of Iterations')
    plt.ylabel('Delta Error')
    return plt
```

```python
# plot iterations vs time
def plot_time(df_stats):
    plt.plot(df_stats['Time'])
    plt.title('Iterations vs Time')
    plt.xlabel('number of Iterations')
    plt.ylabel('Time')
    return plt
```

```python
# plot iterations vs delta error
def plot_value(df_stats):
    plt.plot(df_stats['Max V'], label='max v')
    plt.plot(df_stats['Mean V'], label='mean v')
    plt.title('Iterations vs Max Value & Mean Value')
    plt.xlabel('number of Iterations')
    plt.ylabel('Value')
    plt.legend()
    return plt
```
